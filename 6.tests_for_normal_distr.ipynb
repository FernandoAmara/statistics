#### Tests for normal distribution

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Creating a variable with data in a normal distribution with the rvs function (100 records)
dados = norm.rvs(size = 1000)
dados
```

```python
#histogram
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
# Graphing to check if the distribution is normal
fig, ax = plt.subplots()
stats.probplot(dados, fit=True,   plot=ax)
plt.show()
```

```python
# Shapiro test execution
#second argument is the value of p, we cannot reject the null hypothesis
stats.shapiro(dados)
```

```python
dados2 = skewnorm.rvs(4, size=1000)
```

```python
#histogram
plt.hist(dados2, bins = 20)
plt.title('Dados')
```

```python
# Graphing to check if the distribution is normal
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True,   plot=ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
